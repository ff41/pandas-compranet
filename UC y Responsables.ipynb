# Análisis Responsables y Unidades Compradoras

Este notebook es para documentar el análisis de que una sola persona es responsable de múltiples unidades compradoras.



```python
import pandas as pd
```

Voy a importar el directorio de Unidades Compradoras (UC) en formato csv que se encuentra en el sitio: "https://sites.google.com/site/cnetuc/directorio"

```python
uc = pd.read_csv('Documentos/Pouvoir/QQW/2017/banderas/UC_170612072400.csv')
```

```python
uc.info()
```

```python
uc.RESPONSABLE.value_counts()
```

A partir de estos resultados queremos analizar si las UC en la que los responsables se repiten más de una vez en efecto son distintas.

```python
uc[uc.RESPONSABLE == "AVILA CECEÑA SERGIO"].groupby("NOMBRE_UC").count()
```

Como podemos leer en la columna "NOMBRE_UC" existen dos UC con el mismo nombre pero los cuatro nombres tienen un nombre al final. Podemos comprobar si estas UC tienen una clave única distinta en la columna CLAVE_UC.

```python
uc[uc.RESPONSABLE == "AVILA CECEÑA SERGIO"].groupby("CLAVE_UC").count()
```

En efecto, las cuatro UC están registradas como unidades distintas. Esto implica una bandera por cada responsable que se repita en distintas Unidades Compradoras.

Ahora hay que comprobar que estas UC con sus responsable aparezcan en nuestra BBDD de compranet.

```python
cnet  = pd.read_csv('Documentos/compranet.csv')
```

```python
cnet.info()
```

```python
cnet[cnet.CLAVEUC == "926008990"].groupby("NOMBRE_DE_LA_UC").count()
```

```python
cnet[cnet.CLAVEUC == "926008991"].groupby("NOMBRE_DE_LA_UC").count()
```

```python
cnet[cnet.CLAVEUC == "926008990"].groupby("RESPONSABLE").count()
```

```python
cnet[cnet.CLAVEUC == "926008991"].groupby("RESPONSABLE").count()
```

Aquí se comprueba que la misma persona se encuentra en dos UC distintas. Esto levantaría una bandera a la UC.
