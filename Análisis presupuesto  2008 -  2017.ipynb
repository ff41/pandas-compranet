# Análisis de la información presupuestal presentada en Transparencia Presupuestaria

En el sitio http://www.transparenciapresupuestaria.gob.mx/es/PTP/datos_presupuestarios_abiertos se puede encontrar un csv que contiene datos presupuestarios desde el 2008 hasta el primer trimestre del 2017.
Nos interesa conocer cómo podemos cruzar la información de nuestra bbdd de contrataciones públicas con información presupuestaria.

```python
import pandas as pd
```

```python
pre = pd.read_csv('Descargas/presupuesto_mexico_2008_2017.csv')
```

```python
pre.info()
```

Con esta información podemos conocer cuál ha sido el presupuesto gastado para el proyecto del nuevo aeropuerto para el año 2016.

Primero ubicamos el gasto para el año 2016

```python
pre2016 = pre[pre.CICLO == 2016]
```

```python
pre2016.info()
```

Podemos identificar que existen varias columnas relacionadas con el monto: "MONTO APROBADO", "MONTO EJERCIDO", "MONTO_EJERCICIO", "MONTO_MODIFICADO", "MONTO_DEVENGADO", "MONTO_PAGADO", "MONTO_ADEFAS".
Se puede descargar en la misma página donde se descarga el .csv el diccionario de las columnas del csv. 

El "MONTO_DEVENGADO" es aquel que representa la obligación de pago por la contratación de bienes y servicios por lo que usaremos este número. También conocemos el ID_CLAVE_CARTERA del Nuevo Aeropuerto (NAICM). Esta información la puedes encontrar en la página: https://www.gob.mx/shcp/acciones-y-programas/cartera-publica-de-inversion-16287

```python
pre2016[pre2016.ID_CLAVE_CARTERA == "1409JZL0005"]
```

```python
pre2016NAICM =pre2016[pre2016.ID_CLAVE_CARTERA == "1409JZL0005"]
```

```python
pre2016NAICM.sort_values("MONTO_DEVENGADO", ascending=False).groupby("DESC_UR").MONTO_DEVENGADO.sum()
```

Vemos como la suma del monto devengado "0.0" para el Grupo Aeroportuario de la Ciudad de México, S.A. de C.V. (GACM) Ahora revisamos los partidas específicas que involucra sólo al GACM. Para esto creamos un data frame con sólo filas del GACM.

```python
pre2016NAICM.head(9)
```

Podemos observar que ninguna de las columnas referente al proyecto del nuevo aeropuerto, donde participa el GACM, no tienen datos en ninguna de las columnas de "MONTO". Esto es un error en la base de datos y queda documentado aquí para decidir qué pasos tomar.

Ahora, teniendo como antecedente que no existen datos completos en el proyecto del NAICM. Búsquemos si la Unidad Responsable del nuevo aeropuerto tiene más gastos no relacionados con el NAICM.

```python
pre2016GACM = pre2016[pre2016.DESC_UR == "Grupo Aeroportuario de la Ciudad de México, S.A. de C.V."]
```

```python
pre2016GACM.info()
```

```python
pre2016GACM.ID_CLAVE_CARTERA.value_counts()
```

Aquí se identifica que el Grupo Aeroportuario no tiene catalagodo 100 de sus 102 gastos por lo que necesitamos otro identificador para conocer su monto gastado.
