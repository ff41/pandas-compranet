# Análisis de la información presupuestal presentada en Transparencia Presupuestaria

En el sitio http://www.transparenciapresupuestaria.gob.mx/es/PTP/datos_presupuestarios_abiertos se puede encontrar un csv que contiene datos presupuestarios desde el 2008 hasta el primer trimestre del 2017.
Nos interesa conocer cómo podemos cruzar la información de nuestra bbdd de contrataciones públicas con información presupuestaria.

```python
import pandas as pd
```

```python
pre = pd.read_csv('Descargas/presupuesto_mexico_2008_2017.csv')
```

```python
pre.info()
```

Con esta información podemos conocer cuál ha sido el presupuesto gastado para el proyecto del nuevo aeropuerto para el año 2016.

Primero ubicamos el gasto para el año 2016

```python
pre2016 = pre[pre.CICLO == 2016]
```

```python
pre2016.info()
```

Podemos identificar que existen varias columnas relacionadas con el monto: "MONTO APROBADO", "MONTO EJERCIDO", "MONTO_EJERCICIO", "MONTO_MODIFICADO", "MONTO_DEVENGADO", "MONTO_PAGADO", "MONTO_ADEFAS".
Se puede descargar en la misma página donde se descarga el .csv el diccionario de las columnas del csv. 

El "MONTO_DEVENGADO" es aquel que representa la obligación de pago por la contratación de bienes y servicios por lo que usaremos este número. También conocemos el ID_CLAVE_CARTERA del Nuevo Aeropuerto (NAICM). Esta información la puedes encontrar en la página: https://www.gob.mx/shcp/acciones-y-programas/cartera-publica-de-inversion-16287

```python
pre2016[pre2016.ID_CLAVE_CARTERA == "1409JZL0005"]
```

```python
pre2016NAICM =pre2016[pre2016.ID_CLAVE_CARTERA == "1409JZL0005"]
```

```python
pre2016NAICM.sort_values("MONTO_DEVENGADO", ascending=False).groupby("DESC_UR").MONTO_DEVENGADO.sum()
```

Vemos como la suma del monto devengado "0.0" para el Grupo Aeroportuario de la Ciudad de México, S.A. de C.V. (GACM) Ahora revisamos los partidas específicas que involucra sólo al GACM. Para esto creamos un data frame con sólo filas del GACM.

```python
pre2016NAICM.head(9)
```

Podemos observar que ninguna de las columnas referente al proyecto del nuevo aeropuerto, donde participa el GACM tienen datos en ninguna de las columnas de "MONTO". Esto es un error en la base de datos y queda documentado aquí para decidir qué pasos tomar.

Ahora, teniendo como antecedente que no existen datos completos en el proyecto del NAICM. Búsquemos si la Unidad Responsable del nuevo aeropuerto tiene más gastos no relacionados con el NAICM.

```python
pre2016GACM = pre2016[pre2016.DESC_UR == "Grupo Aeroportuario de la Ciudad de México, S.A. de C.V."]
```

```python
pre2016GACM.info()
```

```python
pre2016GACM.ID_CLAVE_CARTERA.value_counts()
```

Aquí obtenemos los 2 gastos del GACM específicamente para el NAICM

Ahora analicemos estos dos gastos.

```python
naicm = pre2016GACM[pre2016GACM.ID_CLAVE_CARTERA == "1409JZL0005"]
```

```python
naicm.info()
```

```python
naicm.groupby("DESC_PARTIDA_ESPECIFICA").MONTO_DEVENGADO.sum()
```

```python
naicm.groupby("DESC_PARTIDA_ESPECIFICA").MONTO_EJERCICIO.sum()
```

```python
naicm.groupby("DESC_PARTIDA_ESPECIFICA").MONTO_APROBADO.sum()
```

Según los datos del presupuesto en el 2016, en el apartado de la Cartera del proyecto del Nuevo Aeropuerto de la Ciudad de México (NAICM), existieron dos partidas específicas, las cuales no hubo un gasto DEVENGADO ni EJERCIDO aún cuando existía un MONTO APROBADO DE $2,750,000,000.

Sabemos, gracias a la Torre de Control que en el 2016 existieron 69 contratos con actas de fallo de ese mismo año por más de $3,000,000,000 (http://torredecontrol.projectpoder.org/sobre_el_proyecto.html) entre los que se incluyen la construcción del edificio terminal y la acometida elétrica.

Obtenemos dos conclusiones de aspecto general que necesitan ser confirmadas:

- El GACM gastó más dinero que el monto aprobado en el ciclo presupuestado para el proyecto del NAICM del 2016.
- El GACM no justifica los gastos con la Secretaría de Hacienda y Crédito Público ya que los MONTOS de la dependencia en el csv del presupuesto del 2016 no concuerdan.

```python
pre2016GACM.DESC_UR.value_counts()
```

```python
pre2016GACM.ID_CAPITULO.value_counts()
```

```python
pre2016GACM.DESC_CAPITULO.value_counts()
```

Aquí podemos observar los gastos del GACM están repartidos en diferentes capítulos. Según el diccionario, la descripción más específica del gasto proviene de la columna "DESC_PARTIDA_ESPECIFICA".

```python
pre2016GACM.groupby(["DESC_PARTIDA_ESPECIFICA","DESC_CAPITULO"]).MONTO_DEVENGADO.sum().reset_index().sort_values("MONTO_DEVENGADO", ascending=False)
```
